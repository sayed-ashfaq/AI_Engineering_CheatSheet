````python
## GIVEN THE ARRAY SIZE N, FIND COUNT OF ELEMENTS WHICH HAS ATLEAST ONE GREATER ELEMENT THAN ITSELF

def count_elements(arr):
    # subtract count of max elements from len of arr
    # first find the max element
    mx = arr[0]
    count = 0
    for i in arr:
        if i > mx:
            mx = i
    for i in arr[1:]:
        if i == mx:
            count += 1

    return len(arr) - count


print(count_elements([1,2,3,4,5]))
print(count_elements([-3,-2,6,8,4,8,5]))
print(count_elements([2,5,1,4,8,0,8,1,3,8]))
````

### GIVEN AN ARRAY OF N ELEMENTS. FIND COUNT OF PAIR (I, J) WHERE I + J ARE INDICES SUCH THAT

### arr[i] + arr[j] = k where i != j

````python
def finding_sum(arr, k):
    count = 0
    for i in range(len(arr)):
        for j in range(i, len(arr)):
            if arr[i] != arr[j]:
                if arr[i] + arr[j] == k:
                    count += 1
    return count

def count_pairs(arr, k):
    freq = {}
    count = 0

    for x in arr:
        target = k - x
        if target in freq:
            count += freq[target]
        freq[x] = freq.get(x, 0) + 1

    return count

print(finding_sum([1,2,3,4,5], 10))
print(count_pairs([1,2,3,4,5], 5))
````

### Reverse the array

````python
def reverse_slice(arr):
    return arr[::-1]

def reverse(arr):
    for i in range(len(arr)//2):
        arr[i], arr[-i-1] = arr[-i-1], arr[i]
    return arr

print(reverse_slice([-3,-2,6,8,4,8,5, 4]))
print(reverse([-3,-2,6,8,4,8,5]))
````

````python
def reverse_range(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
    return arr

print(reverse_range([2,5,1,4,8,0,8,1,3,8], 2, 6)) # 241808
````

### Given the array of size N. rotate your array k times clockwise.

````python
# input arr, k

def rotate_arr(arr, k, dir= "right"):
    n = len(arr)
    # observe input and expected output and build upon that
    k = k % n # so that k is in range
    if k == 0:
        return arr
    elif dir == "right":
        return arr[-k:] + arr[:-k] # clock wise
    elif dir == "left":
        return arr[k:] + arr[:k]  # anticlock wise


print(rotate_arr([2,5,1,4,8,0,8,1,3,8], 2, dir = "left"))
````

````python
def equilibrium_index(arr):
    n = len(arr)
    prefix_arr = [0]*n
    prefix_arr[0] = arr[0]
    for i in range(1, n):
        prefix_arr[i] = prefix_arr[i-1] + arr[i]

    if prefix_arr[n-1] - prefix_arr[0] == 0:
        return 0
    for s in range(1, n):
        # s is both start and end
        if prefix_arr[s-1] == prefix_arr[n-1] - prefix_arr[s]:
            return s

print("Equilibrium index: ", equilibrium_index([-7,1,5,2,-4,3,0]))

## LEFT SUM AND RIGHT SUM METHOD

def equilibrium_index_(arr):
    n = len(arr)
    total = sum(arr)
    left_sum = 0

    for i in range(n):
        right_sum = total - left_sum - arr[i]
        if left_sum == right_sum:
            return i
        left_sum += arr[i]
    return -1

print("Equilibrium index:", equilibrium_index([-7, 1, 5, 2, -4, 3, 0]))
print("Equilibrium index:", equilibrium_index_([-5,1,-15, 10]))
print("Equilibrium index:", equilibrium_index_([1, 2, 3]))
print("Equilibrium index:", equilibrium_index_([0,-1,1]))

````

#### 2. Split an array into two equal Sum subarrays
Given an array of integers greater than zero, find if it is possible to split it in two subarrays (without reordering the elements), such that the sum of the two subarrays is the same. Print the two subarrays.

````python
# input: arr[] = [1, 2, 3 ,4, 5, 5} ouput: arr = [1, 2, 3][4, 5]

def split_arr_sum(arr):
    total = sum(arr)
    if len(arr) <2:
        return "Empty array, Not Possible"
    left_sum = arr[0]

    for i in range(1, len(arr)):
        right_sum = total - left_sum
        if left_sum == right_sum:
            return arr[:i], arr[i:]
        left_sum += arr[i]
    return "Not Possible"

print(split_arr_sum([1, 2, 3, 4, 5, 5]))
print(split_arr_sum([4, 1, 2, 3]))
print(split_arr_sum([]))
print(split_arr_sum([1,2,3,4,5,6, 1]))

## ANOTHER APPROACH

def split_equal_sum(arr):
    total= sum(arr)

    if total % 2 != 0: # since the sum is odd
        return "Not possible"

    target = total//2
    curr_sum = 0
    for i in range(target):
        curr_sum += arr[i]
        if curr_sum == target:
            left = arr[:i+1]
            right = arr[i+1:]
            return left, right
    return "Not possible"

print("\n====Second Approach===")
print(split_equal_sum([0]))
print(split_equal_sum([1, 2, 3, 3, 3]))
````

#### 3. Mean of range in array

Given an array `arr[]` of n integers and q queries represented by an array `queries[][]`, where `queries[i][0] = l and queries[i][1] = r.` For each query, the task is to calculate the mean of elements in the range l to r and return its floor value.

````python
def mean_of_range(arr, que): # queries are 0 based if 1 based then substract 1 like s= s-1, e= e-1
    n = len(arr)

    prefix_sum = [0]*n
    prefix_sum[0] = arr[0]
    for i in range(1, n):
        prefix_sum[i] = prefix_sum[i-1] + arr[i]

    result = []
    for s, e in que:
        d = e - s + 1
        if s == 0:
            result.append(prefix_sum[e]//d)
        else:
            result.append((prefix_sum[e] - prefix_sum[s-1])//d)
    return result

# example
arr = [1, 2, 3, 4, 5]
queries = [(0, 2), (1, 3), (0, 4)]

print(mean_of_range([3, 7, 2, 8, 5], [[1, 3], [1, 4]]))
print(mean_of_range(arr, queries))
````

#### 4. Original Array from given Prefix Sums
You are given a prefix sum array `presum[]` of an array `arr[]`. The task is to find the original array `arr[]` whose prefix sum is `presum[]`.

````python
def presum_arr(prefix_sum):
    n = len(prefix_sum)
    arr = [prefix_sum[0]]
    for i in range(1, n):
        arr.append(prefix_sum[i] - prefix_sum[i-1])
    return arr

print(presum_arr([5, 7, 10, 11, 18]))
print(presum_arr([45, 57, 63, 78, 89, 97]))

````

#### 5. Product of Array Except Self
Given an array `arr[]` of n integers, construct a product array `res[]` (of the same size) such that `res[i]` is equal to the product of all the elements of `arr[]` except `arr[i]`.

````python
def product_arr(arr):
    n = len(arr)
    res = [1] * n

    # left products
    left = 1
    for i in range(n):
        res[i] = left
        left *= arr[i]

    right = 1
    for i in range(n-1, -1, -1):
        res[i] *= right
        right *= arr[i]

    return res

print(product_arr([10, 3, 5, 6, 2]))
````

#### 2. Given as array of size N. Count the no of leaders in the array,
> Leader: An that element that is strictly greater than all the elements on its right side.

````python
arr = [10, 1, 3, 7, 4, 5] # 10, 7, 5 are leaders so count is 3

def leader_count(arr):

    # add edge cases
    if not arr:
        return 0

    n = len(arr)
    cnt = 1
    mx_ele = arr[-1]
    for i in range(n-2 , -1, -1): # n-2 because you have already included the last element as mx_ele
        if arr[i] > mx_ele:
            cnt += 1
            mx_ele = arr[i]
    return cnt

print(leader_count(arr))
print(leader_count([16, 15, 4, 3, 5, 2]))
````

#### 3. Given an array of size N. Find the length of shortest range which contains both min and max element of array.

````python
def minmax_len(arr):
    n = len(arr)

    #adding edge cases
    if n == 0:
        return 0

    mn = arr[0]
    mx = arr[0]
    for i in range(n):
        if arr[i] < mn:
            mn = arr[i]
        if arr[i] > mx:
            mx = arr[i]

    # adding edge cases
    if mn == mx:
        return 1

    ans = n
    c_min = -1
    c_max = -1
    for i in range(n):
        if arr[i] == mn:
            c_min = i
            if c_max != -1:
                ans = mn(ans, i - c_max + 1)
        elif arr[i] == mx:
            c_max = i
            if c_min != -1:
                ans = mn(ans, i - c_min + 1)

    return ans

# examples
print(minmax_len(arr))
````

What is a Subarray?
Subarray is contigous part of array. It is an array that is part of one big/main array.

- The subarray should maintain the same order as main.
- Complete array OR single element can also be subarray.

How to identify the subarray
- start and end index. start <= end, end <= N

How to calculate possible subarrays in an array
`arr = [4, 5, 3, 1, 9, 12, 24, 5]` of size N.

#### 1. Find total sum of all Subarray sums.

````python
def total_sum_subarr(arr):
    n = len(arr)
    total_sum = 0
    for i in range(n):
        curr_sum = 0
        for j in range(i, n):
            curr_sum += arr[j]
            total_sum += curr_sum
    return total_sum

print(total_sum_subarr([3, 2, -1, 5]))
print(total_sum_subarr([-1, 3, 4]))

## Optimized method
# 1. We don't need all the subarrays but the answer such as 38, 21
# 2. What could be the ways to do that.
# 3. We observed that it's just elements and that are coming x, y , z.. times
# 4. Think how you can get those x, y. z ...
print("===Optimized method===")
## Contribution technique - Contribution of each element in getting the result.
def sum_subarrs(arr):
    n = len(arr)
    total_sum = 0
    for i in range(n):
        total_sum += (i+1)*(n-i)*arr[i]

    return total_sum
print(sum_subarrs([3, 2, -1, 5]))
print(sum_subarrs([-1, 3, 4]))
````
