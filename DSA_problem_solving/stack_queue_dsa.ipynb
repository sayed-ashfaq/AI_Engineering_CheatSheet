# Stack/Queue - DSA

## Stack: LIFO
- stack follows Last In First Out
- Push: Append an element at max index: last index of the array
- pop: Removes the last element in the stack
- top: Gives the last index element, if no element then -1 index
- size: Gives the length of the stack.

## QUeue: FIFO
- Que follows First in First out
- Push: Add elements to the array at the end
- Pop: removes the first element in the array
- top: Give the first index element
- size: gives the current size of array.
