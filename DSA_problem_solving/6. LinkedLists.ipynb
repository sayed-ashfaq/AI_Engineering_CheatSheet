# Linked Lists

#### What is a linked list
- A linked list is a linear data structure where elements are stored as nodes, and each node points/links to the next node.
- It mainly allows efficient insertion and deletion operations. Usefull mainly for Stack, Queue, deque
- Node: A node contains two things first is data and second is link that connects it with another node.
    - The first thing is called the head node which contains value
    - Pointer: The second thing which points to the other node.


**Where is this usefull:**
- Implementing stacks and Queue - Insertion and deletion
- Cache - hashmap or doubly linked list
- Graphs often store negihbors as linked list.

````python
### Getting started

#1. Print a linked list - Singly
class Node():
    def __init__(self, newData):
        self.value = newData
        self.next = None


node1 = Node(3)
node2 = Node(19)
node3 = Node(12)
node4 = Node(21)
node1.next = node2
node2.next = node3
node3.next = node4
````

````python
currentNode = node1
while currentNode:
    print(currentNode.value, end= " -> ")
    currentNode = currentNode.next
print("None")
````

````python
##2. print linkedlist - Doubly
class Node:
    def __init__(self, newData):
        self.value = newData
        self.next = None
        self.prev = None

node1 = Node(4)
node2 = Node(5)
node3 = Node(6)
node4 = Node(7)

node1.next = node2

node2.prev = node1
node2.next = node3

node3.prev = node2
node3.next = node4

node4.prev = node3


print("\nTraversing forward:")
currentNode = node1
while currentNode:
    print(currentNode.value, end=" -> ")
    currentNode = currentNode.next
print("null")

print("\nTraversing backward:")
currentNode = node4
while currentNode:
    print(currentNode.value, end=" -> ")
    currentNode = currentNode.prev
print("null")
````

#### 3. Insert at Head

Write a function to insert a new node at the beginning.

Before

10 → 20 → None


Insert 5

After

5 → 10 → 20 → None

````python
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode

    def display(self):
        curr = self.head
        while curr:
            print(curr.value, end=" -> ")
            curr = curr.next
        print("None")

l1 = LinkedList()

l1.insert_at_beginning(34)
l1.insert_at_beginning(22)
l1.insert_at_beginning(19)

l1.display()
````

#### 4. Insert at Tail

Insert a node at the end.

Before

1 → 2 → None


Insert 3

After

1 → 2 → 3 → None


````python
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode


    def insert_at_end(self, data):
        newNode = Node(data)
        if self.head is None:
            self.head = newNode
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = newNode

    def delete_at_beginning(self):
        if self.head is None:
            return
        else:
            self.head = self.head.next

    def display(self):
        curr = self.head
        while curr:
            print(curr.data, end=" -> ")
            curr = curr.next
        print('None')
````

````python
l1 = LinkedList()
````

````python
l1.insert_at_end(12)
l1.insert_at_end(21)
l1.insert_at_end(19)

l1.display()
````
