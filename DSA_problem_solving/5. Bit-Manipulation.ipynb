## BitWise Operators

Think of bitwise ops as **tools for talking directly to bits** (the language your models, tensors, and hardware *actually* run on).

#### üß† Bitwise Operators (Python Cheat Notes)

##### 1Ô∏è‚É£ AND `&` ‚Üí *Mask / Filter*

```python
a & b
```

* **Rule**: `1 & 1 = 1`, else `0`
* **Use**: keep bits you care about, kill the rest
* **Mnemonic**: *Both must agree*

```python
x = 0b1101
mask = 0b0101
x & mask   # 0b0101
```

---

##### 2Ô∏è‚É£ OR `|` ‚Üí *Force bits ON*

```python
a | b
```

* **Rule**: if **any bit is 1 ‚Üí 1**
* **Use**: enable flags
* **Mnemonic**: *At least one says yes*

```python
x | 0b0010
```

---

##### 3Ô∏è‚É£ XOR `^` ‚Üí *Difference / Toggle*

```python
a ^ b
```

* **Rule**: same ‚Üí 0, different ‚Üí 1
* **Use**: toggling bits, comparisons, hashing
* **Mnemonic**: *Only one survives*

```python
x ^ 0b1111     # flips bits
```

üî• **AI tip**: XOR is common in **hashing, data shuffling, encryption-like ops**

**Example for AND, OR, XOR**
```python
a = 13 # 1101
b = 10 # 1010

a & b =  8  # 1000
a | b = 15 # 1111
a ^ b = 7 # 0111
```

---

##### 4Ô∏è‚É£ NOT `~` ‚Üí *Invert everything*

```python
~a
```

* **Rule**: flips all bits
* **Python gotcha**: infinite 2‚Äôs complement

```python
~5   # = -6
```

Mnemonic:

> `~x = -(x + 1)`

---

##### 5Ô∏è‚É£ Left Shift `<<` ‚Üí *Multiply by powers of 2*

```python
a << n
```

* Shifts bits left, adds zeros
* **Math**: `a * (2**n)`
* **Use**: fast scaling, bit packing

```python

```

Mnemonic: *Shift left ‚Üí grow*

---

##### 6Ô∏è‚É£ Right Shift `>>` ‚Üí *Divide by powers of 2*

```python
a >> n
```

* Shifts bits right
* **Math**: `a // (2**n)`
* **Use**: quantization, compression

```python

```

Mnemonic: *Shift right ‚Üí shrink*

## Bit Manipulation problems

### 1Ô∏è‚É£ Easy ‚Äî Single Number

**Problem**
Given a non-empty array of integers where **every element appears twice except one**, find that single one.

**Example**
`[4,1,2,1,2] ‚Üí 4`

**What this tests**

* XOR properties
* Space optimization

**Expected idea**

* XOR of same numbers = 0
* 0 XOR x = x


````python
arr = [4,1,2,1,2] #4
def UniqueEle(arr):
    ans = 0
    for i in range(len(arr)):
        ans = ans^arr[i]
    return ans

print(UniqueEle(arr))
````

### 2Ô∏è‚É£ Easy‚ÄìMedium ‚Äî Count Set Bits

**Problem**
Given an integer `n`, return the **number of set bits (1s)** in its binary representation.

**Example**
`n = 13 (1101) ‚Üí 3`

**What this tests**

* Bit masking
* Brian Kernighan‚Äôs algorithm

**Constraint**

* Try **without converting to binary string**

````python
def convert_decimal(binary):
    ans = 0
    for i in range(len(binary)):
        ans += int(binary[-1 - i]) * (2 ** i)
    return ans

print("Binary to Decimal Conversion: ",convert_decimal('11010110'))

def ConvertBin(n):
    if n == 0:
        return '0'

    binary_representation = ""
    while n > 0 :
        remainder = n % 2
        binary_representation = str(remainder) + binary_representation
        n = n//2

    return binary_representation


# Brian Kernighan's algorithm
# Since it mentioned do without conversion then

def CalSetbits(n):
    count = 0
    while n > 0:
        n &= (n-1) # n &= n-1
        count += 1
    return count
# n & (n - 1) clears the rightmost set bit (lowest 1) in the binary representation of n.
# 10101101 ‚Üí 10101100 ‚Üí 10101000 ‚Üí 10100000 ‚Üí 10000000 ‚Üí 0

decimal_num = 25

print(f"The binary representation of {decimal_num} is {ConvertBin(decimal_num)}")
print(f"The set bits are {CalSetbits(decimal_num)}")
````

### 4Ô∏è‚É£ Medium‚ÄìHard ‚Äî Two Numbers Appearing Once

**Problem**
Given an array where **every element appears twice except two numbers**, find the two unique numbers.

**Example**
`[1,2,1,3,2,5] ‚Üí [3,5]`

**What this tests**

* XOR partitioning
* Rightmost set bit trick

**Constraint**

* O(n) time
* O(1) extra space

````python
def two_unique_ele(arr): # [1,2,1,3,2,5]
    xor_all = 0
    for n in arr:
        xor_all = xor_all ^ n

    # xor_all gives 6 which is 3 ^ 5

    diff_bit = xor_all & -xor_all
    a, b = 0, 0
    for n in arr:
        if n & diff_bit:
            a ^= n
        else:
            b ^= n
    return a, b

arr = [1,2,1,3,2,5]
print(two_unique_ele(arr))
print(two_unique_ele([5,6,8,8,6,5,7,9]))
````

## Problem Solving from scaler

### Tips and tricks for faster calculation
1. Check ith bit set or unset
- `a & 1 << i` {if 0 False else True}
- `a & 1 << i` converts every thing to 0 except the ith element( 0 if ith ele is 0 else 1).
2. Set the bit at ith position 1 -> 1 , 0 -> 1.
- `a | 1 << i` { converts the ith element to 1 }
3. Toggle the ith bit 1 -> 0, 0 -> 1
- `a ^ 1 << i` if same element then 0, if 1 in any converts to 1.
4. Unset the bit at ith position 1 -> 0, 0 -> 0
- **use if condition**:`if a & 1 << 0 != 0: then a ^ 1<<i`

These are the four occurrences that should remember at all costs.


````python
# 1. given number n, find ith bit set or unset. -- CHECK

def set_unset(n, i):
    d = 1 << i
    if n & d ==0:
        return False
    return True


print(set_unset(214, 6)) # '11010110'
````

````python
## 2. Set the ith bit. ans: OR and Left shift
def to_binary(n):
    return format(n, 'b')


def set_ith_bit(n, i):
    return n | (1 << i)

number = int("10101101", 2)
bit_position = 4

print(f"Original binary value : {to_binary(number)}")
print(f"Original decimal value: {number}")

updated = set_ith_bit(number, bit_position)

print(f"\nSetting bit at position {bit_position}")
print(f"Updated binary value  : {to_binary(updated)}")
print(f"Updated decimal value : {updated}")
````

````python
## 3. Toggle the ith bit (XOR and Left shift)

def toggle_bit(n, i):
    return n ^ (1 << i)

def to_binary(n):
    return format(n, 'b')

number = int("10101101", 2)
bit_position = 5

print(f"Original binary value : {to_binary(number)}")
print(f"Original decimal value: {number}")

updated = toggle_bit(number, bit_position)

print(f"\nToggle bit at position {bit_position}")
print(f"Updated binary value  : {to_binary(updated)}")
print(f"Updated decimal value : {updated}")
````

````python
## 4. unset the ith bit - use conditional XOR

def unset(n, i):
    if n & (1 << i) != 0:
        return n ^ (1 << i)
    return n

def to_binary(n):
    return format(n, 'b')

number = int("10101101", 2)
bit_position = 1

print(f"Original binary value : {to_binary(number)}")
print(f"Original decimal value: {number}")

updated = unset(number, bit_position)

print(f"\nToggle bit at position {bit_position}")
print(f"Updated binary value  : {to_binary(updated)}")
print(f"Updated decimal value : {updated}")
````
