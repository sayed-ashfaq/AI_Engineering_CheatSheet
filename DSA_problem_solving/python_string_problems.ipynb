```python
def to_uppercase(s):
    result = ""
    for char in s:
        if 'a' <= char <= 'z':
            result += chr(ord(char) - 32)
        else:
            result += char
    return result

print(to_uppercase("Python"))
print(to_uppercase("somerandome_1+2"))
```

### LowerCase

```python
def to_lowercase(s):
    result = ""
    for char in s:
        if "A" <= char <= "Z":
            result += chr(ord(char) + 32)
        else:
            result += char
    return result

print(to_lowercase("PYTHON"))
print(to_lowercase("SOMERANDOMESTRING+2"))
```

### First Non-repeating char

```python
def is_first_non_repeating(s):
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1
    for i in range(len(s)):
        if char_count[s[i]] == 1:
            return i
    return -1

print(is_first_non_repeating("sswawiss"))
print(is_first_non_repeating("oasisa"))
print(is_first_non_repeating("aaaaaa"))
```

### is palindrome or not

```python
# simple method using slicing
def is_palindrome(s):
    if s == s[::-1]:
        return True
    return False


print(is_palindrome("maddam"))
```

```python
# loops, two pointer technique

def is_palindrome_tp(s): # madam
    i = 0
    j = len(s) - 1 # j = 5-1 = 4 indexing
    while i < j: # 0 < 4
        # compare first and last letter
        # m = m if not return false
        if s[i] != s[j]:
            return False
        i += 1
        j -= 1
        return True

print(is_palindrome_tp("maddam"))
print(is_palindrome_tp("hello"))
print(is_palindrome_tp("forexerof"))
```

### Find the most frequent character in a string

```python
def frequent_char(s):
    s = s.replace(' ', '')
    count_char = {}
    for char in s:
        if char in count_char:
            count_char[char] += 1
        else:
            count_char[char] = 1
    max_char = ""
    max_count = 0
    for char, count in count_char.items():
        if count > max_count:
            max_char = char
            max_count = count

    return max_char, max_count


print(frequent_char("hello"))
print(frequent_char("sayedashfaq"))
print(frequent_char("This is very very bad"))
```

### Compressed String

```python
def compressed_str(s): # aaabbccc -> a3b2c2
    compressed_result = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            count += 1
        else:
            compressed_result.append(s[i-1]+str(count))
            count = 1
    compressed_result.append(s[-1]+str(count))

    compressed_str = "".join(compressed_result)
    if len(compressed_str) < len(s):
        return compressed_str
    return s

print(compressed_str("aaabbccc"))
print(compressed_str("qqccdd"))
```

### Find all permutations of a string - recursion

```python
# abc = [abc, bac, cba, cab,....] # straight function: itertools.permutations()

# Hardcoded solution

def find_permutations(s): # abc
    #base
    if len(s) == 1:
        return [s]
    permutations = []
    for i in range(len(s)):
        current_char = s[i] # a
        remaining_char = s[:i] + s[i+1:] #bc
        remaining_permutations = find_permutations(remaining_char)  # bc, ca
        for p in remaining_permutations:
            permutations.append(current_char+p)
    return permutations

print(find_permutations("abc"))
print(find_permutations("ab12"))


```

### Roman to integer conversion

```python
# RULES IN ROMAN NUMBERING SYSTEM
# {I : 1, V : 5, X : 10, L : 50, C : 100, D : 500, M : 1000}

## 1. Larger to smaller: VI -> add -> 6

## 2. Smaller to Larger: IV -> subtract -> 4

def roman_to_num(s):
    roman_dict = {'I': 1, 'V' : 5, 'X' : 10, 'L' : 50,
                  'C' : 100, 'D' : 500, 'M' : 1000}
    prev_value = 0
    total= 0

    for letter in s[::-1]: # go in reverse

        curr_value = roman_dict[letter]

        if curr_value >= prev_value:
            total += curr_value
        else:
            total -= curr_value

        prev_value = curr_value

    return total

print(roman_to_num("MMDCCCXCIV"))
```

### Return length of the longest substring

```python
# Find the length of the longest string where each char is repeated atleast k times

# divide and conquer, sliding window, recursion

def longest_substr(s, k):
    # base
    if len(s) < k:
        return 0

    # char count
    char_count = {}
    for char in s:
        if char in char_count:
            char_count[char] += 1
        else:
            char_count[char] = 1

    # max count
    for char in char_count:
        if char_count[char] < k:
            max_count = 0
            for part in s.split(char):
                max_count = max(max_count, longest_substr(part, k))
            return max_count
    return len(s)

print(longest_substr('aaabbc', 3))
print(longest_substr('ababaabcbc', 3))
print(longest_substr('abdddcedcece', 3))
```

### Longest Substring Without Repeating Characters

```python
# two pointers, sliding window

def longest_substring(s): #abcabca
    left = 0 # l_index
    seen = {} # store char: index_location
    max_length = 0
    for right in range(len(s)):
        char = s[right] # r_index
        if char in seen and seen[char] >= left:
            left = seen[char] + 1

        seen[char] = right
        max_length = max(max_length, right-left+1)

    return max_length

# Example
print(longest_substring("abcabcdbb"))  # 3
print(longest_substring("bbbbb"))     # 1
print(longest_substring("cadbzabcd"))    # 5
```

```python
def convert(s):
    result = ''
    for char in s:
        if char >= 'a' and char <= 'z':
            result += chr(ord(char) - 32)
        if char >= "A" and char <= "Z":
            result += chr(ord(char) + 32)
    return result

print(convert("aAbBcC"))
```
