## Understanding MAPE
- MAPE : Mean absolute percentage Error
- It is a statistical metric for evaluating the accuracy of forecasting models.
- It expresses prediction errors as percentages easier to interpret and compare.

MAPE = (1/n) * Σ(|actual – forecast| / |actual|) * 100


```
import numpy as np

def mape(actual, forecast):
    actual, forecast = np.array(actual), np.array(forecast)
    return np.mean(np.abs((actual - forecast)/actual))
```

```
actual = [11, 13,6, 5]
forecast = [5, 6.8, 8.66, 7.862]
print(mape(actual, forecast))
```

## ACF and pACF

### ACF - Auto Correlation Factor
- AutoCorrelation measures the degree of similarity/correlation between given times series and the lagged version of that time series over successive time periods. Included all the datapoitns from to start to end.
- Measures indirect relationship




- The PACF at LAG 0 is 1.0. This is always the case. A value is always 100% correlated with itself!
    - At lag zero ACF is always equal to one, which makes sense because the signal is always perfectly correlated with itself.
- For an MA(q) model, the acf will be zero at all lags beyond q, but the MA(q) can be written as an AR(infinity). Therefore, the pacf will never be zero, but will decline geometrically.
- For an AR(p), however, once the effects of y(t-1), y(t-2), …, y(t-p) are removed, the correlation between yt and y(t-p-j) will be zero for all positive integer values of j.
    - So, whilst the acf for an AR(p) will decline geometrically, the pacf will be zero after p lags.

Finally, although the acf will be zero at lag 2 for an MA(1), the pacf will not.
