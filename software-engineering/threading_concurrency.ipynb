## Threading and Multi Processing
Run Code concurrently using the Threading Module
- When do we use Threading?
    - We use threading when ever it is going to speed up our program.
    - it can come from running multiple things concurrently.

```python
import time

start = time.perf_counter() # more like a stopwatch to know how much time it took to run this code
print("start: ", start)
def do_something():
    print("sleeping 1 second ....")
    time.sleep(1)
    print("Done Sleeping")

do_something()
do_something()
finish = time.perf_counter()
print("finish: ", finish)
print(f"Finished in {finish - start} seconds")
```
