# Day 2 Problem: Rate-Limited Event Aggregator

```python
events = [
    (100, "u1", "click"),
    (120, "u1", "click"),      # ignored (within 60s)
    (170, "u1", "view"),       # valid

    (200, "u2", "click"),
    (220, "u2", "purchase"),  # ignored
    (260, "u2", "purchase"),  # valid

    (90,  "u3", "view"),
    (140, "u3", "view"),      # ignored
    (200, "u3", "view"),      # valid

    (50,  "u4", "click"),
    (110, "u4", "click"),     # valid (exactly 60s gap)
]
```

```python
def aggregate_events(events: list[tuple]) -> dict:
    # Sort events by time to correctly apply rolling window logic
    events.sort(key=lambda x: x[0])

    last_event_time = {}      # user_id -> last accepted event_time
    counts = {}               # event_type -> count

    for event_time, user_id, event_type in events:
        if user_id not in last_event_time or event_time - last_event_time[user_id] >= 60:
            last_event_time[user_id] = event_time
            counts[event_type] = counts.get(event_type, 0) + 1

    # return the aggregated event dict
    return counts
aggregate_events(events)
```
